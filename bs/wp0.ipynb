```python
# import modules
import requests
from bs4 import BeautifulSoup
```

```python
# download the HTML content of the URL
page = requests.get("https://en.wikipedia.org/wiki/Protein")
 
# display status
print('was the page downloaded successfully? %s' % page.status_code)
 
# display scraped data
print(page.content)
```

```python
# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')
 
# display scraped data
print(soup.prettify())
```

```python
list(soup.children)
 
# find all occurrence of p in HTML
# includes HTML tags
print(soup.find_all('p'))
 
print('\n\n')
 
# return only text
# does not include HTML tags
print(soup.find_all('p')[0].get_text())
```

```python
object = soup.find(id="mp-left")
 
# find tags
items = object.find_all(class_="mp-h2")
result = items[0]
 
# display tags
print(result.prettify())
```
