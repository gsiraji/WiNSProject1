```python
import igraph as ig

import numpy as np

import scipy
import random 
random.seed(0)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
G = ig.Graph(
n = 4,

edges = [(0,1),
        (0,2),
        (0,3),
        (1,2),
        (1,3),
        (2,3)],
directed = False)

G.summary()
```

```python
n = G.vcount()
m = G.ecount()

print(f'{n} nodes, {m} edges')
```

```python
G = ig.Graph.Famous('Zachary')
```

```python
G['layout'] = G.layout_fruchterman_reingold()
G['layout'].coords
```

```python
G.vs['color'] = 'blue'
G.vs['size'] = 0.4
G.es['color'] = 'grey'
fig, ax = plt.subplots()
ig.plot(G, target = ax, layout =G['layout'])
ax.set_title('Zachary Graph')
```

```python
vs = G.vs[[0, 1, 2, 3]]
es = G.es.select(_within=vs)
vs['color'] = 'green'
es['color'] = 'green'
fig, ax = plt.subplots()
ig.plot(G, target = ax, layout =G['layout'])
```

```python
neighbors = G.neighbors(4)
print(neighbors, G.degree(4))
```
