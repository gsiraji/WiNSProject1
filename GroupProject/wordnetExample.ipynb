```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```python
def get_lemmas(word):

    lemmasList = [str(lemma.name()) for lemma in wn.synset(word).lemmas()]
    return lemmasList
```

```python
# look up  a word using synsets() and use pos argument to specify the part of speech

print(wn.synsets('book', pos=wn.NOUN))
```

```python
print(wn.synset('book.n.03').examples())

print(wn.synset('soup.n.03').definition())


print(get_lemmas('book.n.03'))
```

```python
# get the synonyms of a word

book = wn.synset('book.n.01')

book.hypernyms()
```
